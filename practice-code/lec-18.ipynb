# File casestudy

```python
import os
```

```python
directory = "res"
```

```python
csv_filename = os.path.join(directory, "datafile.csv")
```

```python
csv_filename
```

```python
rows = []

with open(csv_filename, 'r') as f:
   for line in f:
       print (line.strip().split(','))
       rows.append(line.strip().split(','))
```

```python
rows
```

```python
for r in rows[1:]:
    for i in range(2, 6):
        r[i] = float(r[i])
```

```python
rows
```

```python
'|'.join(['a', 'b', 'c'])
```

```python
out_rows = []

for r in rows:
    r = [str(c) for c in r]
    out_rows.append(r)
```

```python
out_rows
```

```python
for r in out_rows:
    out_line = ','.join(r)
    print (out_line)
```

```python
csv_filename_out = os.path.join(directory, 'datafile-out.csv')
```

```python
csv_filename_out
```

```python
with open(csv_filename_out, 'w') as f:
    for r in out_rows:
        out_line = ','.join(r)
        f.write(out_line + '\n')
```
