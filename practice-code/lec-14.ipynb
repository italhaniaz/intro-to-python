```python
def make_tea(ing):
    for i in ing:
        print ("Put", i, "in")

ingredients = ["sugar", "milk", "tea leaves"]
make_tea(ingredients)
```

```python
def count(seq, value):
    """Counts the value in the sequence of seq."""
    counter = 0
    
    for i in seq:
        
        if i == value:
            counter = counter + 1

    return counter
```

```python
s = [1, 2, 3, 4, 1, 4, 3, 4, 4, 5, 5, 7]
count(s, 2)
```

```python
count(s, 1)
```

```python
count(s, 4)
```
