```python
def preassure(v, t, n):
    """
    Computer the preassure in pascals

    Applies the ideal gas law.

    v -- volume of gas in cubic meter
    t -- absolute temperature in degrees
    n -- particals of gas
    """
    k = 1.38 - 23
    return n * k * t / v
```

```python
def absolute_value(x):
    """ Calculate the absolute value (x)
    x -- it is any real number.
    """

    if x > 0:
        return x
    elif x == 0:
        return 0
    else:
        return -x
```
