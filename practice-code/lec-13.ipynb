# Fibonacci Numbers

```python
def fib(n):
    if n <= 1:
        return n
    else:
        return fib(n-1) + fib(n-2)
```

```python
a = fib(3)
```

```python
def fib(n):
    if n <= 1:
        return n

    a, b = 0, 1

    for i in range(1, n + 1):
        n = a + b
        a = b
        b = n

    return a
    
```

```python
%time fib(40)
```

```python
def fib(n):
    if n <= 1:
        return n

    a, b = 0, 1

    for i in range(1, n):
        a, b = b, a + b

    return b
    
```

```python
%time fib(50)
```
