# Lists and Mutability

```python
x = 34
y = x
```

```python
print (f"{x}, {y}")
```

```python
y = 20
print (f"{x}, {y}")
```

```python
mammals = ['cat', 'kangaroo', 'horse']
fish = ['tune', 'shark', 'catfish']
```

```python
animals = [mammals, fish]
```

```python
print (animals)
```

```python
mammals[0] = 'mule'
```

```python
print (mammals)
print (animals)
```

```python
mammals_2 = mammals
mammals_2[0] = 'dog'
print (mammals)
print (animals)
```

```python
mammals_3 = mammals[:]
```

```python
mammals_3[0] = 'mouse'
print (mammals_3)
print (mammals)
print (animals)
```

```python
s1 = ['student 1', 'Ali', 'hamza']
s2 = ['student 2', 'talha', 'niaz']
s3 = ['student 3', 'afzal', 'yaseen']
```

```python
students = [s1, s2, s3] 
```

```python
students_copy = students[:]
```

```python
students_copy[0][0] = 'sudent 1 changed'
```

```python
print (students)
print (students_copy)
```

```python
print(s1)
```

```python
import copy
```

```python
student_3 = copy.deepcopy(students)
```

```python
student_3[0][0] = 'student 1 is changed'
print (student_3)
print (students)
```

```python
def change_var(x):
    x = x + 23

def change_list(lst):
    lst.append(24)
```

```python
x = 20
change_var(x)

print (x)
```

```python
lst = [100, 54, 32]

change_list(lst)
print (lst)
```
