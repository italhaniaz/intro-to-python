# File system

```python
directory = 'res'
name = 'dummy.txt'
```

```python
import os
```

```python
filename =  os.path.join(directory, name)
```

```python
filename
```

```python
f = open(filename, 'r')

for line in f:
    print (line)

f.close()  #always remember this.
```

```python
f = open(filename, 'r')

for line in f:
    print (line.strip())

f.close()  #always remember this.
```

# **Best Practice**

```python
with open(filename, 'r') as f:
    for line in f:
        print (line.strip())
```

```python
f
```

```python
with open(filename, 'r') as f:
    data = f.read()
```
