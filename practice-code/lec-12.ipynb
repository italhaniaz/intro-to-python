```python
def is_even(n):
    if n % 2 == 0:
        return True
    else:
        return False


def is_odd(n):
    if n % 2 != 0:
        return True
    else:
        return False

for i in range(2, 10):

    if not is_odd(i):
        continue   # go back to while loop.
        
    print (i)

print ("done")
```

```python
for i in range(2, 10):

    if i % 5 == 0:
        break

    print (i)

print ("done")
```

```python
def factorial(n):

    fac = 1
    i = 1
    while i <= n:
        fac = fac * i
        i = i + 1

    print (fac)
```

```python
factorial(6)
```

```python
def fac(n):
    fac = 1
    for i in range(1, n):
        fac = fac * i

    print (fac)
```

```python
fac(7)
```
